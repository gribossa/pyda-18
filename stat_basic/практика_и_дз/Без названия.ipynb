```python
import pandas as pd
import numpy as np
```

```python
horses_df = pd.read_csv('/Users/sonyasha/pyda_homeworks/stat_basic/практика_и_дз/horse_data.csv', na_values=['?','-'])
```

```python
horses_df.head(5)
```

```python
horses_df = horses_df[['2', '1', '38.50', '66', '28', '3', '5', '2.2']]
```

```python
horses_df.rename(columns={'2': 'surgery', '1': 'age', '38.50': 'rectal temperature', '66': 'pulse', '28': 'respiratory rate', '3': 'temperature of extremities', '5': 'pain', '2.2': 'outcome'}, inplace=True)
```

```python
horses_df.head(5)
```

```python
horses_df.describe()
```

```python
print('Значение моды для параметра наличие операции: ', horses_df['surgery'].mode()[0])
print('Значение моды для параметра возраст: ', horses_df['age'].mode()[0])
print('Значение моды для параметра боль: ', horses_df['pain'].mode()[0])
print('Значение моды для параметра итог: ', horses_df['outcome'].mode()[0])
print('Значение моды для параметра температура конечностей: ', horses_df['temperature of extremities'].mode()[0])
```

Данные значения свидетельствуют о том, что большинство лошадей в предоставленных данных это возрастные лошади, при лечении которых применялось хирургическое вмешательство, кроме того, большинство пациентов испытывали болезненные ощущения среднего уровня и после лечения продолжили жить свою нормальную лошадиную жизнь

```python
def outliers(df):
    """
    Рассчитывает выбросы для каждого столбца DataFrame на основе межквартильного размаха
    """
    for column in df:
        Q1 = horses_df[column].quantile(0.25)
        Q3 = horses_df[column].quantile(0.75)
        IQR = Q3 - Q1
        low_outlier = Q1 - (1.5*IQR)
        high_outlier = Q3 + (1.5*IQR)
        print(f'In сolumn {column}: lower_outlier is {low_outlier}, higher_outlier is {high_outlier}')
```

```python
outliers(horses_df[['rectal temperature', 'pulse', 'respiratory rate']])
```

 Выводы по выбросам:
 1. surgery это харктеристика наличия или отсутствия операции, имеющая значения либо 1 либо 2, согласно прикрепленному описанию, соотвественно любой выброс по характеристике означает ошибку ввода данных
 2. то же самое касается характеристики age, так как согласно описанию, лошадей делят по возрасту на молодых и взрослых, давая в соответствие цифру 1 или 2. Во время выгрузки оказалось, что цифра 2 в данных отсутствует, но встречается 9, предполагаю, что это ошибка ввода и 9 означает 2
 3. в столбце rectal temperature выбросы носят скорее всего, естественный харктер, так как заведомо известно, что это данные о болеющих лошадях, и повышенная (или пониженная) температура у них это нормально, и это подтверждается тем, что согласно подсчетам средняя температура по предоставленным данным выше, чем нормальная температура, указанная в описании к данным
 4. по столбцу pulse неадекватное значение нижней границы объясняется слишком большим разбросом между минимальным и максимальным значениями и большим количеством повышенных значений температуры, что логично приводит к увеличению межквартильного размаха. Согласно описанию, нижняя граница нормы сердченого ритма в спокойном состоянии 20 ударов, что даже меньше минимального значения по столбцу pulse. Выбросы же в большую сторону можно объяснить сильными болезненными ощущениями у конкретных лошадей, скорее всего они так же являются естественными
 5. в столбце respiratory rate неадекватные значения нижней границы обусловлены так же, как в пульсе, слишком большим размахом значений. В описании указано, что между значениями по данному параметру большие колебания, что вполне соответствует приведенным данным, и означает, что скорее всего выбросы являются естественными
 6. в столбце temperature of extremities значения являются категориальными, так что любой выброс из дискретных значений от 1 до 4, является ошибкой ввода данных
 7. то же самое можно сказать и о столбцах pain и outcome, однако, согласно max и min значениям в таблице, таких ошибок там нет.

Так как в столбцах surgery и outcome не хватает всего по одному значению, а все остальные значения категориальные, заменяем их на моду.

```python
horses_df['outcome'].fillna(horses_df['outcome'].mode()[0], inplace=True)
horses_df['surgery'].fillna(horses_df['surgery'].mode()[0], inplace=True)
```

В столбце пульс не хватает всего 24 значений, что представляет недостаток всего в 8%, можно попытаться заполнить на основе медианного значения, так как разброс данных слешком большой 

```python
horses_df['pulse'].fillna(horses_df['pulse'].median(), inplace=True)
```

По приложенному описанию можно предположить, что значения пульса коррелируют с уровнем боли, к тому же для млекопитающих характерен более высокий пульс в юном возрасте. На основании этого, считаю уместным заполнять пробелы по уровню боли значениями, соответствующими возрастным категориям с определенным уровнем сердечного ритма

```python
horses_df['pain'].fillna(horses_df.groupby(['age', 'pulse'])['pain'].transform('median'), inplace=True)
```

Так как в описании к данным указано, что respiratory rate не используется для анализа из-за больших колебаний значений, то можно заменить все NaN средними значениями для поддержания общей статистики

```python
horses_df['respiratory rate'].fillna(horses_df['respiratory rate'].mean(), inplace=True)
```

Остаются значения ректальной температуры и температуры конечностей. Имеет смысл брать для ректальной температуры среднее значение, тк среднее и медиана близко, а для температуры конечностей моду, так как это категория и мода равна медиане.

```python
horses_df['temperature of extremities'].fillna(horses_df['outcome'].mode()[0], inplace=True)
horses_df['temperature of extremities'].fillna(horses_df['outcome'].mean(), inplace=True)
```
