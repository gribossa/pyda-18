```python
ratings = pd.read_csv('/Users/sonyasha/pyda_homeworks/pandas_2.0/доп_материалы_к_пандас_1_и_2/ml-latest-small/ratings.csv')
ratings
```

```python
ratings_pivot = ratings.pivot_table(index = 'userId', columns = 'rating', values = 'timestamp', aggfunc = 'count', fill_value = 0, 
                    margins = True)
ratings_pivot
```

```python
userId_list = list(ratings_pivot.loc[ratings_pivot.All > 100].index)
userId_list.remove('All')
```

```python
time_of_life = 0
times = []
counts = 0

for userId_ in userId_list:
    time_of_life = time_of_life + (ratings.loc[ratings.userId == userId_].timestamp.max() - ratings.loc[ratings.userId == userId_].timestamp.min())
    counts += 1
    
mean_time = time_of_life / counts
```

```python
round(mean_time, 2)
```
